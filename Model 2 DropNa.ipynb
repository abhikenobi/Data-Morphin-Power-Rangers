### K Nearest Neighbors (Ned Flanders)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
y = model_df["Potability"]
target_names = ["non-potable", "potable"]
```

```python
X = model_df.drop("Potability", axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler model and fit it to the training data

X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 30, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    

    
plt.figure(figsize=[13,8])    
plt.plot(range(1, 30, 2), train_scores, label = 'Testing Accuracy')
plt.plot(range(1, 30, 2), test_scores, label = 'Training Accuracy')
plt.legend()
plt.title("K Nearest Neighbors")
plt.xlabel("# of Neighbors")
plt.ylabel("Testing Accuracy Score")
plt.savefig('./Resources/Model2/kneighbors.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_scores),1+test_scores.index(np.max(test_scores))))
```
